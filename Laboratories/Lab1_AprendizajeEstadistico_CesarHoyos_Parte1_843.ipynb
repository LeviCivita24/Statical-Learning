**9. Normalizar cada columna respecto a su media y desviación estandar**

$$ \frac{x - mean(x)}{std(x)} $$

````python
#@title Solucion Punto 9
# tomemos la media y la desviación estandar de cada columna
df2_mean = df2.mean()
df2_std = df2.std()
# Se toma un nuevoDataframe que tome solo las columnas sobre las cuales se puede realizar la operacion numerica
columnasSeleccionadas = df2.columns[1:-1] # Se crea una mascara
DF = df2[columnasSeleccionadas] # Se aplica la mascara al DataFrame

k1 = 0 # Se inicializa un contador
for i in DF: # Se recorre cada columna
  DF[i] = (DF[i] - df2_mean[k1]) / df2_std[k1] # Sobre todos los elementos de la columna se hace la operacion
  k1 += 1

newDf1 = pd.concat([df2['Diagnosis'], DF], axis=1) # Se utiliza el concat para volver a formar el DataFrame original
df3 = pd.concat([newDf1, df2['DiagnosisNumeric']], axis=1)
df3.tail()
````

**12. Determinar los datos outlier para la columna RadiusMean y eliminarlos del DataFrame, para ello se debe construir un gráfico tipo boxplot**

En estadística un outlier es un dato que difiere significativamente de las otras observaciones. Esto puede ser consecuencia a la variabilidad en la medida, una indicación de nuevos datos, o el resultado de un error experimental. Este tipo de datos puede causar un gran error a la hora de hacer un analisis estadistico.

````python
df3.boxplot(column="RadiusMean", by='Diagnosis', sym = 'k.', figsize=(18,6))
````

Como se evidencia de los boxplots se puede observar a simple vista que tanto para las etiquetas B y M se tienen tres datos outliers.

````python
#@title Solucion Punto 12 - Parte1
Q1 = df3['RadiusMean'].quantile(0.25) # Se toma el quartil al 25%
Q3 = df3['RadiusMean'].quantile(0.75) # Se toma el quartial al 75%
IQR = Q3 - Q1
print(Q1)
print(IQR)
df3 = df3[~(df3['RadiusMean'] < (Q1 - 1.5 * IQR))]
df3 = df3[~(df3['RadiusMean'] > (Q3 + 1.5 * IQR))]
df3.boxplot(column="RadiusMean", by='Diagnosis', sym = 'k.', figsize=(18,6))
````

El método que se trato de utilizar fue el de Turkey el cual está basado en las medidas de los rangos intercuantiles. El cual define que si Q1 y Q3 son los cuartiles más bajos y altos, respectivamente. Se puede definir una observación outlier como:

$$[ Q_1 - K(Q_3 - Q_1), Q_3 + K (Q_3 - Q_1)]$$

Generalmente se recomienda tomar un valor para K = 1.5

````python
#@title Solucion Punto 12 - Parte 2
df3 = df3[~(df3['RadiusMean'] < (Q1 - 0.8 * IQR))]
df3 = df3[~(df3['RadiusMean'] > (Q3 + 0.2 * IQR))]
df3.boxplot(column="RadiusMean", by='Diagnosis', sym = 'k.', figsize=(18,6))
````

Como se evidencia en el desarrollo anterior, la formula de Turker es muy útil para quitar los outliers en el caso de cancer maligno no es tan util en el caso de cancer benigno. Tocaría entrar a entender a mayor profundidad otras fórmulas para extrar los outliers para determinar una conclusión certera.

**13. Encontrar la matriz de correlación, emplear el método corr(), dentro de seaborn buscar el método heatmap() para realizar este tipo de gráficos.**

````python
#@title Solucion Punto 13
# Calcula la matriz de correlación
correlation_matrix = df3.corr()

# Crea una figura y un eje utilizando matplotlib
plt.figure(figsize=(8, 6))

# Crea un mapa de calor utilizando Seaborn
sns.heatmap(correlation_matrix, cmap='coolwarm', fmt=".2f")

# Título del gráfico
plt.title('Matriz de Correlación')

# Mostrar el gráfico
plt.show()
````

La matriz de correlación, es una matriz que representa la relación entre las diferentes variables. Los coeficientes de correlación miden cuán fuerte están relacionadas las variables. Esto también puede ser usado para encontrar situaciones donde dos o más variables están altamente correlacionadas con otra. Esto es llamado multicolinealidad.

Si se observa la gráfica anterior la variable RadiusMean con PerimeterMean están fuertemente correlacionados, lo que implica que un cambio en una de ellas realiza un cambio en la otra. Sin embargo, RadiusMean con SmoothnessSe tienen una correlación negativa, lo que puede implicar que mientras una de ellas crece la otra decrece. También, se pueden tener valores de correlación 0, lo que indica que el cambio de una variable no afecta la otra variable.
