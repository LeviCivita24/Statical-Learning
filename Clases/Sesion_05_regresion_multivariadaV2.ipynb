```python
N = 100
x_1 = 2 * np.random.random(N) - 1
x_2 = 2 * np.random.random(N) - 1

y_label = 0.2 * x_1 - 0.5 * x_2 #+ 0.2*(2*np.random.random(N)-1)
df = pd.DataFrame({"x1":x_1, "x2":x_2, "y":y_label})
df["ones"]=1
```

```python
X = df[["ones", "x1", "x2"]].values.T
Y = df.y.values.reshape(1, N)
```

```python
def model(theta, X):
  h=theta.T@X
  return h

def cost(h, Y):
  J = (h-Y)**2
  return 0.5*J.mean()

def grad(learning_rate,theta,  h, Y, X):
  theta = theta - (learning_rate * np.dot((h-Y), X.T)).T
  return theta
```

```python
class linearRegresion():
  def __init__(self, X, Y):
    features=np.shape(X)[0]
    self.theta =  np.random.random(features).reshape(features, 1)
    self.X = X
    self.Y = Y
    self.H =  self.theta.T@self.X
    self.J = []

  def model(self):
   self.H = self.theta.T@self.X

  def cost(self):
    J1 = (self.H-self.Y)**2
    self.J.append(0.5*J1.mean())

  def update_theta(self,learning_rate):
    self.theta = self.theta - (learning_rate * np.dot((self.H-self.Y), self.X.T)).T

  def fit(self, learning_rate):
    for i in range(0, 100):
      self.model()
      self.cost()
      self.update_theta(learning_rate)
```

```python
N = 100
x_1 = 2*np.random.random(N)-1
x_2 = 2*np.random.random(N)-1

y_label = 0.2 * x_1 - 0.5 * x_2 #+ 0.01*(2*np.random.random(N)-1)
df = pd.DataFrame({"x1":x_1, "x2":x_2, "y":y_label})
df["ones"] = 1

X = df[["ones", "x1", "x2"]].values.T
Y = df.y.values.reshape(1, N)

lr = linearRegresion(X, Y)
lr.fit(0.01)
lr.theta
```
